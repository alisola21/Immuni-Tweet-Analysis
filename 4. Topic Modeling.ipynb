```python
from bertopic import BERTopic
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
```

```python
df = pd.read_csv("dati/df_finale_2020_2021_nltk_analysis.csv")
del df['Unnamed: 0']
```

```python
df_filter = df.loc[(df["likes_count"]>=1) & (df["retweets_count"]>= 1)]
df = df_filter
df.shape
```

```python
df2020_= df[df["date"] <= "2020-12-31"]
df2020_.shape
```

```python
df2021_= df[df["date"] >= "2021-01-01"]
df2021_.shape
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
tweets = df.tweet.to_list()
timestamp = df.date.to_list()
```

```python
tweets2020 = df2020_.tweet.to_list()
timestamp2020_ = df2020_.date.to_list()
```

```python
tweets2021 = df2021_.tweet.to_list()
timestamp2021_ = df2021_.date.to_list()
```

```python
vectorizer_model = CountVectorizer(ngram_range=(1, 2), stop_words = nltk.corpus.stopwords.words('italian'))
topic_model = BERTopic(vectorizer_model=vectorizer_model, language="italian", nr_topics = 10) #20 argomenti più in voga
topics, probs = topic_model.fit_transform(tweets)
```

```python
vectorizer_model = CountVectorizer(ngram_range=(1, 2), stop_words = nltk.corpus.stopwords.words('italian'))
topic_model2020 = BERTopic(vectorizer_model=vectorizer_model, language="italian", nr_topics = 10) #20 argomenti più in voga
topics, probs = topic_model2020.fit_transform(tweets2020)
```

```python
vectorizer_model= CountVectorizer(ngram_range=(1, 2), stop_words = nltk.corpus.stopwords.words('italian'))
topic_model2021 = BERTopic(vectorizer_model=vectorizer_model, language="italian", nr_topics = 10) #20 argomenti più in voga
topics, probs = topic_model2021.fit_transform(tweets2021)
```
